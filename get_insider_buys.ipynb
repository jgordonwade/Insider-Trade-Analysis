* Purpose of this notebook: Read Ingest insider trading data from the SEC and produce a dataframe with salient information.

* Input is a list of ticker symbols, via a file `tickerlist.csv`
* Output is a dataframe `insider_buys.csv`

This file  `insider_buys.csv` is then used in the notebook `insider_trade_detection`

```python
import insider_trade_detector as itd
import yfinance as yf
import time
```

```python
symbol_list = itd.read_tickerlist_csv('tickerlist_naz100.csv')
#symbol_list = ['AAPL', 'MSFT', 'INTC']
start_date, late_ticker = itd.get_start_date(symbol_list, start_date_pad=30)
print(len(symbol_list), 'tickers. Start date: ', start_date.strftime('%Y-%m-%d'), '\t "Latest" ticker is', late_ticker)
```

#### Loop to get relevant SEC data.

The main function here is `ticker_to_insider_buys(ticker)`

    - **Input:** A string containing a single  ticker symbol.

    - **Output:** A list (`insider_buys`), where each element is a dictionary containing relevant information about an insider buy transaction. These data include: accession number, ticker, trade and filing dates, number of shares, price per share, insider type, and the URL of the XML filing.

Then *save this info to disk* via the function `update_insider_buys_file`. The data is saved in `'insider_buys.csv'`


```python
overall_start_time = time.time()
insider_buys = []
#for tikr in symbol_list[:0]:    # <==-- Remove the ":0" here to run this loop. But it takes about 2-3 hours!
for tikr in symbol_list:
    buys_list = itd.ticker_to_insider_buys(tikr)
    insider_buys += buys_list
    itd.update_insider_buys_file(buys_list)
    print(tikr, 'done. \n')
overall_end_time = time.time()
print((overall_end_time-overall_start_time)/60, "minutes")
```
