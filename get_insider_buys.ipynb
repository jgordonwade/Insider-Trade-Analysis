* Purpose of this notebook: Read Ingest insider trading data from the SEC and produce a dataframe with salient information.

* Input is a list of ticker symbols, via a file `tickerlist.csv`
* Output is a dataframe `insider_buys.csv`

This file  `insider_buys.csv` is then used in the notebook `insider_trade_detection`

```python
import insider_trade_detector as itd
import yfinance as yf
import time
import csv
```

```python
symbol_list = itd.read_tickerlist_csv('tickerlist.csv')
symbol_list = [s for s in symbol_list if s != 'RIVN']
symbol_list = [s for s in symbol_list if s != 'GFS']
symbol_list = [s for s in symbol_list if s != 'ABNB']
```

```python
#symbol_list = ['RIVN','TSLA']
start_date, late_ticker = itd.get_start_date(symbol_list, start_date_pad=30)
print(len(symbol_list), 'tickers. Start date: ', start_date.strftime('%Y-%m-%d'), '\t "Latest" ticker is', late_ticker)
earliest_date = start_date.strftime('%Y-%m-%d')
```

#### Get relevant SEC data.

The main function here is `tikcker_list_to_insider_buy_list(symbol_list,earliest_date)`

```python
insider_buy_list = itd.tikcker_list_to_insider_buy_list(symbol_list,earliest_date)
```

```python
csv_file = 'insider_buys.csv'
with open(csv_file, mode='w', newline='') as file:
    writer = csv.DictWriter(file, fieldnames=insider_buy_list[0].keys())
    
    # Write the header
    writer.writeheader()
    
    # Write the rows
    for row in insider_buy_list:
        # Convert insider_type list to a string
        row['insider_type'] = ', '.join(row['insider_type'])
        writer.writerow(row)

print(f"Data written to {csv_file}")
```
